# A Stochastic (random) variable
Consider a stochastic (random) variable $X$. It is capable of "realizing" (ie. taking on, yielding, retunring, etc.) any value within a range of values. Additionally it is non deterministic meaning that it is capable of generating different values assuming that multiple observations are made. A classical example used to describe a random variable is a fair-sided coin or fair-sided die. When we flip a coin or roll the dice, we dont know exactly what value will be realized. We try to explain our expecation for what values can be observed and how likely those observations would be using a probability distribution.

$$ X = \{x_i\} $$

# A Stochastic (random) Process

A process is a collection or set of variables that are indexed by some term $i$. 

$$ X_i = \{X_i\}_{i \in I} $$

The value of the random variable $X_i$ coresponding for a particular value $i$ of the index $I$ is $X_i$ which varies or changes depending on the index.

It is also common to consider a process as a function that considers an index

$$ f(i) = X_i $$

There are many exampls of stochastic processes that were developed to fit certain use cases or applications. We will have a look at a few of these examples.

A stochastic process is commonly referred to as a random walk as it is a mathematical model which represents the idea of a person walking in random directions.

We will see that stochastic processes are defined in several ways: sometimes as a collection of properties which must be satisfied, somtimes as an equation which exhibits a collection of properties, sometimes as a partial differential equation.

## Properties
Stochastic processes can exhibit any number of mathematical properties. Below we will examine a number of common properties.

### Increments

A process $X_t$ can be thought of as a linear combination of increments $\Delta_i$. The increment being the change between realized values. The following statements describe this concept:

$$ X_i = \Delta_0 + \Delta_1 + \cdots \Delta_i$$

$$ \Delta_i = X_i - X_{i-1} $$

$$ X_i = X_{i-1} + \Delta_i$$

### Stationarity

A process is considered stationary if the the mean, variance and autocorrelation structure do not change over time. While the values realized by the process itself may change relative to the index, the the statistics of the process are constant, they do not move, and thus they are stationary.

This invariance implies that the process increments are independent and the realized values are independent.

In addition to a process, it is also common to speak about stationarity in terms of the increments of a process which we will cover later.

To help conceptualize this stationarity, consider the following graph which illustrates a random walk for a unidimensional random variable.

We can see the graph on top looks like noise while the bottom graph looks like it has a more deliberate or time-dependent trend.

The simplest example of a stationary process would be rolling a dice. No matter how many times you roll, no matter the previous roll, the probability of a given outcome is always the same. We look at this in more detail below.

There are several types of stationary which we will examine below.

#### Strong Stationarity

Mathematically put, the probability distribution of the process $X_t$ should be invariant with respect to time (ie. it does not consider time as a parameter) and thus:

$$ F_X(x_1, \cdots x_n) = F_X(x_1, \cdots x_n, t) $$

For more information see [wikipedia](https://en.wikipedia.org/wiki/Stationary_process).

#### Weak Stationarity
Weak stationary is far less restrictive than the strong counterpart. It only requires that the mean and autocorrelation are constant while the variance and other moments can vary and are expected to be finite.

For more information see [wikipedia](https://en.wikipedia.org/wiki/Stationary_process).

#### Stationarity Of Increments

Some processes are characterized by properties being applied to the increments of the process. It is common for a process to be defined as having stationary increments.

$$ X_i = \Delta_0 + \Delta_1 + \cdots \Delta_i$$

$$ \Delta_i = X_i - X_{i-1} $$

$$ \Delta_i \sim \mathcal{D} $$

Wher $\mathcal{D}$ is an arbitrary distribution.

In some texts (like [this](https://en.wikipedia.org/wiki/L%C3%A9vy_process) and [this](https://galton.uchicago.edu/~lalley/Courses/385/LevyProcesses.pdf) the stationarity of increments is expressed as:

<center>$X_t - X_s $ is equal in distribution to $X_{t-1} - X_0$</center>

###  Noise Process
As noted before, a random process is represented as a collection of realized events that are indexed. A classic example of an index is time. In other words, our process is a mathematical model for a variable indexed through (one that changes over) time. 

As with the study of probability we will start by considering scenarios that produces finite, deiscrete, independent events through time such as: flipping a coin three times, rolling a dice six times, choosing the winning horse 10 times in a row at a horse race. 

If we stick with the example of rolling a dice, mathematically, We can begin to develop a model for this process which is expressed algebraically as:

$$ X_t \sim \mathcal{U}(1,6)$$

Here we see that $X_t$ follows a uniform distribution $\mathcal{U}$ with an interval $[1,6]$.

Note: The index $t$ is commonly used to denote and distinguish time as the dimension for the index.

Below we see a graph representing dice rolls over time

```python
from matplotlib import pyplot
import numpy

numpy.random.seed(42)
x = numpy.linspace(0, 100)
y = [int(numpy.random.uniform(1, 7)) for i in x]

# Generate potting objects
px = 1/pyplot.rcParams['figure.dpi']  # pixel in inches
fig, ax = pyplot.subplots(figsize=(600*px, 400*px))
ax.plot(x,y)
ax.set_xlabel('t', size=16)
ax.set_ylabel('$x_t$', size=20)
ax.set_title("Rolling a Dice vs. Time", pad=20 , size=16)
pyplot.show()
```

We can see here that this process is index-independent, or memory-less, or non-recursive. The value at any given point does not depend on the values observed for prior values of the index. Put another way, the probabilitiy is always the same regardless of how many times we have rolled.

This type of random process is commonly referred to as **noise**. There are many types of noise based on the distribution. For example guassian white noise is a stochastic process which follows the standard gaussian distribution.

It is also referred to as a **stationary** process as one of its only properties is that it is stationary.

While this is a helpful model in some cases, in others we need a more sophistocated model which is not index-independent. In other words, the probailities of an event will change based on the prior events.

### Levy Process

#### Overview

A Lévy processes is a stochastic process with independent and stationary stationary increments.

The most well known examples of Lévy processes are the Wiener process, often called the Brownian motion process, and the Poisson process. Further important examples include the Gamma process, the Pascal process, and the Meixner process. 

#### History
The process was named after the French mathematician Paul Lévy,

#### Definition

A Lévy processes $\mathcal{L}_t$ has the following characteristics:

1. $\mathcal{L}_0 = 0$ almost surely
2. Independence of increments: $\mathcal{L}_i \perp \mathcal{L}_j$
3. Stationary increments $\mathcal{L}_i - \mathcal{L}_j \sim \mathcal{D}$
4. The probability space of $\mathcal{L}$ is continuous



### Weiner Process (Standard Brownian Motion)

A methematical model was constructed to explain phenominon. In a series of papers starting in 1918, Weiner constructed the first rigorous model for a random process that exhibited the specified characteristics. 

Weiners work focused on a special case of the brownian motion. As we will see the Weiner process is characterized by standard increments (ie. increments that follow a standard normal distribution). For this reason the weiner process is sometimes referred to as a standard brownian motion.

### Brownian Motion

#### Overview

#### Definition
The brownian motion $\mathcal{B}_t$ is a levy process characterized by the following additional properties:

1. $\Delta_t \sim \mathcal{N}(\mu t, \sigma^2 t)$

This property links the brownian motion to a gaussian process as $\Delta_t$ is a gausian process.

It also links the brownian motion with the weiner process as the brownian motion can be defined as:

$$ \mathcal{B}_t = \mu t + \sigma \mathcal{W}_t $$

Where $\mu$ is referred to as a drift parameter while $\sigma$ is referred to as the scale or noise parameter.

This leads to the realization that the weiner process is a standardized brownian motion. We see the familiar z-score formula (kernel):

$$ \mathcal{W}_t = \frac{\mathcal{B}_t - \mu t}{\sigma}$$


#### References

- https://www.randomservices.org/random/brown/Drift.html

For example consider the brownian motion defined formally as a vairble $X$ inddexed through time $t$.

$$ X_t = \mu t  + \sigma \mathcal{W_t} $$

In the graph below we see the result of constructing an experiment where we observe the brownian motion over time.

```python
from scipy.stats import norm
from matplotlib import pyplot
import numpy

# Process parameters
delta = 0.5
dt = 0.1
n = 20

# Initial condition.
y_0 = 0.0

# Setup data containers
x = numpy.arange(0, 100)
y = numpy.array([y_0])

# Iterate to compute the steps of the Brownian motion.
numpy.random.seed(seed=42)
for i in range(1, len(x)):
    y = numpy.append(y, y[i - 1] + norm.rvs(scale=delta**2*dt))
    
# Generate potting objects
px = 1/pyplot.rcParams['figure.dpi']  # pixel in inches
fig, ax = pyplot.subplots(figsize=(600*px, 400*px))
ax.plot(x,y,color='blue')
ax.set_ylim([-0.3, 0.3])
ax.set_xlabel('t', size=16)
ax.set_ylabel('$x_t$' , size=20)
ax.set_title("Single realization of the random process", pad=20, size=16)
pyplot.show()
```

# Distribution Of Functions

As a stochastic process is random and thus the observations will not be deterministic; ie. we will hot have the same realizations between experiments or trials. In the graph below we see such a situation. We conduct multiple experiments to observe the realizations of the same process on separate occasions.

```python
# Setup data containers
for j in range(1, c):
    y = numpy.vstack([y, numpy.zeros(x.shape[0])])

# Iterate to compute the steps of the Brownian motion.
numpy.random.seed(seed=42)
for i in range(1, len(x)):
    for j in range(1, c):
        y[j][i] = y[j][i - 1] + norm.rvs(scale=delta**2*dt)
    
# Generate potting objects
colors = ['blue', 'red', 'orange', 'purple', 'green', 'yellow']
px = 1/pyplot.rcParams['figure.dpi']  # pixel in inches
fig, ax = pyplot.subplots(figsize=(600*px, 400*px))
for j in range(0, c):
    ax.plot(x,y[j], color=colors[j])
ax.set_ylim([-0.3, 0.3])
ax.set_xlabel('t', size=16)
ax.set_ylabel('$x_t$', size=26)
ax.set_title("Multiple realizations of the random process", pad=20, size=16)
pyplot.show()
```

As we can see, for a particular value of the index $t$, say $t=10$ as shown below, there are multiple values $x_t$ that may be observed.

```python
from scipy.stats import norm
from matplotlib import pyplot
import numpy
    
# Generate potting objects
px = 1/pyplot.rcParams['figure.dpi']  # pixel in inches
fig, ax = pyplot.subplots(figsize=(600*px, 400*px))
ax.axvline(x[10],c='black')
for j in range(0, c):
    ax.plot(x,y[j], color=colors[j])
    ax.scatter(x[10], y[j][10], color=colors[j])
ax.set_ylim([-0.3, 0.3])
ax.set_xlabel('t', size=16)
ax.set_ylabel('$x_t$', size=26)
ax.set_title("Multiple realizations of the random process", pad=20, size=16)
pyplot.show()
```

As such, we can treat the situation as a random variable that realizes a process rather then a point of a process.

Because $X_{t=10}$ is a realization of a random variable, there is a distribution that describes the probability of realizing the given value of $x_t$ for the given point $t=10$.

```python
points_x = [x[10] for i in y]
points_y = [i[10] for i in y]
points_mu = numpy.mean(points_y)
points_sigma = numpy.std(points_y)
rv = norm(loc=points_mu, scale=points_sigma)
p_x = numpy.arange(-.3, .3, .01)
p_y = rv.pdf(p_x)

px = 1/pyplot.rcParams['figure.dpi']  # pixel in inches
fig, ax = pyplot.subplots(figsize=(600*px, 400*px))
ax.plot(p_x, p_y, color='black')
for i in range(0, c):
    point_px = y[i][10]
    point_py = rv.pdf(point_px)
    ax.scatter(point_px, point_py, color=colors[i])

ax.set_xlim([-.3,.3])

pyplot.show()
```

With this mental framework, we can consider each one of these paths (ie. the experiment where we observed the process), as a separate and distinct function. Labeling our graph we can see the distinction.
