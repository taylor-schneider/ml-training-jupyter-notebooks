# Graphing with Matplotlib
In this notebook we will explore how the matplotlib can be used to create common graphical representations of data using common data science libraries.

- In part 1 we look at scatter plots
- In part 2 we look at line graphs

# Part 1 - Examples of creating scatter plots

#### 1.1 : Scatter plot 2D data - using two separate arrays

```python
# Import numpy array
import numpy

# make up some data
x = [numpy.datetime64("2019-01-01"),
     numpy.datetime64("2019-01-02"),
     numpy.datetime64("2019-01-03"),
     numpy.datetime64("2019-01-04"),
     numpy.datetime64("2019-01-05")]

y = [5, 10, 15, 13, 20]

# Import the libraries to plot data
import matplotlib.pyplot as plt
%matplotlib inline

# Create a new graph object
figure, axis = plt.subplots()

# Set the dimensions of the graph area
axis.set_xlim(left=numpy.datetime64("2018-12-31"), right=numpy.datetime64("2019-01-06"))

# Graph a scatter plot on the graph we created
plt.scatter(x,y)

# Beautify the x-labels
plt.gcf().autofmt_xdate()

# Show the plot
plt.show()

```

#### 1.3: Scatter plot 2D data - using 2D ndarray

```python
# Import numpy array
import numpy

# make up some data (x, y)
my_ndarray = numpy.array([[numpy.datetime64("2019-01-01"), 5],
                         [numpy.datetime64("2019-01-02"), 10],
                         [numpy.datetime64("2019-01-03"), 15],
                         [numpy.datetime64("2019-01-04"), 20],
                         [numpy.datetime64("2019-01-05"), 25]])

# Select the data 
x = my_ndarray[:, 0].tolist()    # If we dont convert to a list, an exception will be raised
y = my_ndarray[:, 1]

# Import the libraries to plot data
import matplotlib.pyplot as plt
%matplotlib inline

# Create a new graph object
figure, axis = plt.subplots()

# Set the dimensions of the graph area
axis.set_xlim(left=numpy.datetime64("2018-12-31"), right=numpy.datetime64("2019-01-06"))

# Graph a scatter plot on the graph we created
plt.scatter(x,y)

# Beautify the x-labels
plt.gcf().autofmt_xdate()

# Show the plot
plt.show()
```

# Example 1.5: Scatter plot 3D data using 3D numpy ndarray

```python
# https://stackoverflow.com/questions/42677160/matplotlib-3d-scatter-plot-date
import numpy

# Import our ploting library (included with matplotlib package)
import matplotlib.pyplot as plt, numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Make up some data (in triplet form of [x, y, z])
# where the variables represent date, price, and option
my_ndarray = numpy.array([[numpy.datetime64("2019-01-01"), 5, 23],
                         [numpy.datetime64("2019-01-02"), 10, 30],
                         [numpy.datetime64("2019-01-03"), 15, 34],
                         [numpy.datetime64("2019-01-04"), 20, 40],
                         [numpy.datetime64("2019-01-05"), 25, 50]])

# Select the data 
x = [x for x in range(1,6)]
y = my_ndarray[:, 1].tolist()
z = my_ndarray[:, 2].tolist()

# Import the libraries to plot data
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure('scatter dates')
ax = fig.add_subplot(111, projection='3d')
dates = my_ndarray[:, 0]

ax.scatter(x, y, z)
ax.xaxis.set_ticks(x)
ax.xaxis.set_ticklabels(dates)
plt.show()
```
